```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
import sys

sys.path.append("../../utils")

from shannon import calc_coper
from adjust_molecule import get_bond_angle, adjust_bond_angle
from convert import ob_2_pyscf
```

```python
def plot_bond_angle(mol_name, angle_min, angle_max, num_points, idx_1=2, idx_2=1, idx_3=3, basis='sto-3g', xc='B3LYP'):
    x = np.linspace(angle_min, angle_max, num=num_points)
    y = np.zeros(len(x))

    for i in range(len(x)):
        print(i)
        mol_ob = adjust_bond_angle(mol_name, idx_1, idx_2, idx_3, x[i])
        y[i] = calc_coper(mol_ob, basis=basis, xc=xc)
        
    plt.xlabel("Bond angle (Â°)")
    plt.ylabel("CoPer")
    plt.axvline(x=get_bond_angle(mol_name, idx_1, idx_2, idx_3), color='r')
    plt.plot(x, y)
    plt.savefig(str.format("plots/bond_angle/{mol}_bondangle_{min}_{max}_{num}.png", mol=mol_name, min=angle_min, max=angle_max, num=num_points))
    plt.close()
```

```python
plot_bond_angle(mol_name='CH4', angle_min=0.01, angle_max=180, num_points=1000)
```

Plot BH<sub>3</sub>, CH<sub>4</sub>, NH<sub>3</sub>, H<sub>2</sub>O,  bond angles (symmetric)

```python
mols = ['BH3', 'H2O', 'NH3', 'CH4']
for mol in mols:
    plot_bond_angle(mol, angle_min=0.01, angle_max=180, num_points=1000)
```
